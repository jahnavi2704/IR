# Import statements

```python
import mwparserfromhell 
import subprocess, bz2
from timeit import Timer
import xml.sax
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import subprocess
import json

data_path = '/Users/appletest/IR_project/datasets/enwiki-20191101-pages-articles-multistream.xml.bz2'
books = [] # Every element of the list is the infobox of books as a tuple
writers = [] # Every element of the list is the infobox of writers as a tuple
```

# Code for content handler for XML

```python
class WikiXmlHandler(xml.sax.handler.ContentHandler):
    """Content handler for Wiki XML data using SAX"""
    def __init__(self):
        xml.sax.handler.ContentHandler.__init__(self)
        self._buffer = None
        self._values = {}
        self._current_tag = None
        self._pages = []
        self._article_count = 0

    def characters(self, content):
        """Characters between the opening and closing tags"""
        if self._current_tag:
            self._buffer.append(content)

    def startElement(self, name, attrs):
        """Opening tag of every element"""
        if name in ('title', 'text'):
            self._current_tag = name
            self._buffer = []


    def endElement(self, name):
        """Closing tag of every element"""
        if name == self._current_tag:
            self._values[name] = ' '.join(self._buffer)
        if name == 'page':
            self._article_count += 1
            # The selected page is sent to the process article function
            book = process_article(**self._values, 
                                   template = 'Infobox book')
            writer = process_article(**self._values, 
                                   template = 'Infobox writer')
            # If the article found is a book append to the list of books
            if book:
                 print("book ---> " + book[0])
                 books.append(book)
            # If the article found is a writer append to the list of writers
            if writer:
                print("writer ---> " + writer[0])
                writers.append(writer)
```

```python
def process_article(title, text, template):
    """Looks a wikipedia article for required template"""
    
    # To create a parsing object using mwparserfromhell
    wikicode = mwparserfromhell.parse(text)
    
    # Search through templates for the template
    matches = wikicode.filter_templates(matches = template)
    
    if len(matches) >= 1:
        # Extraction of information from infobox found
        properties = {param.name.strip_code().strip(): param.value.strip_code().strip() 
                      for param in matches[0].params
                      if param.value.strip_code().strip()}
        
        return (title, properties)
```

# Sample output of the names of the infoboxes retrieved

```python
# Creating an Object for handling XML
handler = WikiXmlHandler()
# Parsing the object created
parser = xml.sax.make_parser()
parser.setContentHandler(handler)
start = Timer()
# Parsing the file
for line in subprocess.Popen(['bzcat'], 
                             stdin = open(data_path), 
                             stdout = subprocess.PIPE).stdout:
    try:
        parser.feed(line)
    except StopIterati4on:
        break
    
end = Timer()
```

# Code to convert the obtained list of tuples in json format 

```python
file = open("books.xml", "a")
file.write("<add>" + "\n")


count = 0  # Count variable to assign a unique id to each infobox entity


for book in books:
    file.write("<doc>" + "\n")
    count = count + 1
    file.write("\t" + "<field name=" + "\"" + "id" + "\"" + ">" + str(count) + "</field>" + "\n")
    for attributes in book[1]:
        file.write("\t" + "<field name=" + "\"" + attributes + "\"" + ">" +  str(book[1][attributes]) + "</field>" + "\n" )
    file.write("</doc>" + "\n")
file.write("</add>")
file.close()


file1 = open("writers.xml", "a")
file1.write("<add>" + "\n")
count = 0
for writer in writers:
    file1.write("<doc>" + "\n")
    count = count + 1
    file1.write("\t" + "<field name=" + "\"" + "id" + "\"" + ">" + str(count) + "</field>" + "\n")
    for attributes in writer[1]:
        file1.write("\t" + "<field name=" + "\"" + attributes + "\"" + ">" +  str(writer[1][attributes]) + "</field>" + "\n" )
        #print(attributes)
    file1.write("</doc>" + "\n")
file1.write("</add>")
file1.close()
```

# Question answering system

```python
# The information is obtained only when the attributes are one among these
attributes_author = ["image", "caption", "pseudonym", "birth_place", "spouse", "birth_date" "death_place", "resting_place", "occupation", "nationality", "period", "language", "genre", "movement", "notableworks", "relatives", "signature", "awards" "_version_"]
attributes_book = ["image", "caption", "author", "country", "language", "genre", "published"]
```

# Code for stop word elimination & other unimportant terms from the query

```python
def Preprocess_query(query):
    b = ""
    a = query
    for k in a.split("\n"):
        b = b + re.sub(r"[^a-zA-Z0-9]", ' ', k)


    example_sent = b
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(example_sent)

    filtered_sentence = [w for w in word_tokens if not w in stop_words] 

    filtered_sentence = [] 

    for w in word_tokens: 
        if w not in stop_words: 
            filtered_sentence.append(w) 

    return filtered_sentence   # The filtered query containing important terms
```

# Code for retrieving answers from the Solr database

```python
def Answer_retrieval(filtered_sentence):
    for i in filtered_sentence:
        command = 'curl "http://localhost:8983/solr/Core_writers/select?indent=on&q=name:' + i + "\""
        p = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        out, err = p.communicate()
        out = out.decode("utf-8")
        res = json.loads(out)
        if res["response"]["numFound"] is not 0:
            for j in attributes_author:
                if j is "birth_place":
                    k = "birthplace"
                elif j is "death_place":
                    k = "deathplace"
                elif j is "resting_place":
                    k = "restingplace"
                elif j is "birth_date":
                    k = "birthdate"
                else:
                    k = j
                if k in filtered_sentence:
                    try:
                        print(res["response"]["docs"][0][j])
                    except:
                        pass


    for i in filtered_sentence:
        command1 = 'curl "http://localhost:8983/solr/Core_book/select?indent=on&q=name:' + i + "\""
        p1 = subprocess.Popen(command1, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        out1, err1 = p1.communicate()
        out1 = out1.decode("utf-8")
        res1 = json.loads(out1)
        if res1["response"]["numFound"] is not 0:
            for j in attributes_book:
                if j in filtered_sentence:
                    try:
                        print(res1["response"]["docs"][0][j])
                    except:
                        pass
```

# Enter queries : 

```python
for i in range(5):
    print("Enter your query:")
    query = input()
    filtered_sentence = Preprocess_query(query)
    print()
    print("The filtered query : ", filtered_sentence)
    Answer_retrieval(filtered_sentence)
    print()
```
